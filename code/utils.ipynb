```python
from PIL import Image
```

```python
me = Image.open('../assets/images/me.jpg')
w, h = me.size
print(w, h)
```

```python
trim = (w - h) // 2
trim
```

```python
me_cropped = me.crop((trim, 0, w - trim, h))
me_cropped.size
```

```python
me_small = me_cropped.resize((240, 240))
me_small
```

```python
me_small.save('../assets/images/me_small.jpg')
```
